```python
import os
from dotenv import load_dotenv
from fredapi import Fred

load_dotenv() 

api_key = os.getenv("FRED_API_KEY")
if not api_key:
    raise RuntimeError(
        "FRED_API_KEY not found. Create a .env file or set an environment variable."
    )

fred = Fred(api_key=api_key)
wti = fred.get_series("MCOILWTICO")
copper = fred.get_series("PCOPPUSDM")
coffee = fred.get_series("PCOFFOTMUSDM")


copper.dropna(inplace=True)
coffee.dropna(inplace=True)


wti.to_csv("wti_prices.csv", header=["WTI_price"], index_label="Date")
copper.to_csv("copper_prices.csv", header=["Copper_price"], index_label="Date")
coffee.to_csv("coffee_prices.csv", header=["Coffee_price"], index_label="Date")


```
